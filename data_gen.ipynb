```python
def gen_rn(g_l):
    lst_fn = []
    lst_ln = []
    for i in g_l:
        if i == None:
            lst_fn.append(None)
            lst_ln.append(None)
        elif i == 'M':
            lst_fn.append(names.get_first_name(gender='male'))
            lst_ln.append(names.get_last_name())
        else:
            lst_fn.append(names.get_first_name(gender='female'))
            lst_ln.append(names.get_last_name())
    return lst_fn, lst_ln
```

```python
def ran_un(fn_lst, ln_lst):
    letters = string.ascii_lowercase
    usr_lst = []
    for f, l in zip(fn_lst, ln_lst):
        usrn = ''
        if f == None:
            l4 = ''.join(random.choice(letters) for i in range(4))
            usrn = '{}{:04d}'.format(l4, random.randrange(1, 9999))
            while usrn in usr_lst: usrn = '{}{:04d}'.format(l4, random.randrange(1, 9999))
        else:
            l4 = '{}_{}'.format(f[0].lower(), l.lower())
            usrn = '{}{:04d}'.format(l4, random.randrange(1, 9999))
            while usrn in usr_lst: usrn = '{}{:04d}'.format(l4, random.randrange(1, 9999))
        usr_lst.append(usrn)
    return usr_lst
```

```python
def trans(lst):
    tr_list = []
    for v in lst:
        if v == None:
            tr_list.append(None)
        else:
            tr_list.append(np.random.choice(['bicycle', 'walking', 'car'], p=[0.5, 0.4, 0.1]))
    return tr_list
```

```python
def occup(lst):
    occ_lst = []
    for v in lst:
        if v == None:
            occ_lst.append(None)
        else:
            occ_lst.append(np.random.choice(['student', 'part-time', 'full-time', 'elder', 'jobless'], p=[0.3, 0.2, 0.3, 0.1, 0.1]))
    return occ_lst
    
```

```python
def gen_usrdata(n):
    gen_lst = np.random.choice(['M', 'F', None], size=n, p = [0.25, 0.25, 0.5])
    fn_lst, ln_lst = gen_rn(gen_lst)
    un_lst = ran_un(fn_lst, ln_lst)
    oc_list = occup(gen_lst)
    tr_list = trans(gen_lst)
    data = {'username': un_lst, 'password': ['psw']*n, 'first_name': fn_lst, \
                     'last_name': ln_lst, 'gender':gen_lst, 'occupation': oc_list, 'tr_mode':tr_list}
    df_data = df(data, columns = ['username', 'password', 'first_name', 'last_name', 'gender', 'occupation', 'tr_mode'])
    return df_data
```

```python
def gen_proddata(arr):
    veg = arr[0] 
    fruit = arr[1]
    meat = arr[2]
    vc = arr[3] 
    fc = arr[4]
    mc = arr[5]
    t, sc1, sc2, sc3, sc4, n1, n2, n3, n4 = [], [], [], [], [], [], [], [], []
    for i in range(len(veg)):
        t.append('V')
        sc1.append(veg[i][0].decode('utf'))
        sc2.append(veg[i][1].decode('utf') if len(veg[i]) > 1 else None)
        sc3.append(veg[i][2].decode('utf') if len(veg[i]) > 2 else None)
        sc4.append(veg[i][3].decode('utf') if len(veg[i]) > 3 else None)
        n1.append(veg[i][0].decode('utf'))
        n2.append(veg[i][1].decode('utf') if len(veg[i]) > 1 else None)
        n3.append(veg[i][2].decode('utf') if len(veg[i]) > 2 else None)
        n4.append(veg[i][3].decode('utf') if len(veg[i]) > 3 else None)
    for i in range(len(fruit)):
        t.append('F')
        sc1.append(fruit[i][0].decode('utf'))
        sc2.append(fruit[i][1].decode('utf') if len(fruit[i]) > 1 else None)
        sc3.append(fruit[i][2].decode('utf') if len(fruit[i]) > 2 else None)
        sc4.append(fruit[i][3].decode('utf') if len(fruit[i]) > 3 else None)
        n1.append(fruit[i][0].decode('utf'))
        n2.append(fruit[i][1].decode('utf') if len(fruit[i]) > 1 else None)
        n3.append(fruit[i][2].decode('utf') if len(fruit[i]) > 2 else None)
        n4.append(fruit[i][3].decode('utf') if len(fruit[i]) > 3 else None)
    for i in range(len(meat)):
        t.append('M')
        sc1.append(meat[i][0].decode('utf'))
        sc2.append(meat[i][1].decode('utf') if len(meat[i]) > 1 else None)
        sc3.append(meat[i][2].decode('utf') if len(meat[i]) > 2 else None)
        sc4.append(meat[i][3].decode('utf') if len(meat[i]) > 3 else None)
        n1.append(meat[i][0].decode('utf'))
        n2.append(meat[i][1].decode('utf') if len(meat[i]) > 1 else None)
        n3.append(meat[i][2].decode('utf') if len(meat[i]) > 2 else None)
        n4.append(meat[i][3].decode('utf') if len(meat[i]) > 3 else None)
    mat =  [0] * len(sc1)
    for k, v in vc.items():
        mat[n1.index(k)] = v
    for k, v in fc.items():
        mat[n1.index(k)] = v
    for k, v in mc.items():
        mat[n1.index(k)] = v
    mat = np.array(mat)
    data = {'product_name': n1, 'pack_c': mat[:, 0], 'pack_nc': mat[:, 1], 'single_item': mat[:, 2],\
            'weight':mat[:, 3], 'type':t, 'category_name': sc1, 'pna1': n2, 'pna2': n3, 'pna3': n4,\
            'cna1': sc2, 'cna2': sc3, 'cna3':sc4}
    df_data = df(data, columns = ['product_name', 'category_name', 'type', \
                                  'pna1', 'pna2', 'pna3', 'cna1', 'cna2', 'cna3'])
    df_data_fc = df(data, columns = ['product_name', 'category_name', 'type', 'pack_c', 'pack_nc', 'single_item',\
                                  'weight', 'pna1', 'pna2', 'pna3', 'cna1', 'cna2', 'cna3'])
    return df_data, df_data_fc
```

```python
def gen_storedata(shop):
    shop = [[arr[0].decode('utf'), arr[1].decode('utf'), arr[2], arr[3]] for arr in shop]
    mat = np.array(shop)
    data = {'shop_name':mat[:,0], 'brand':mat[:,1], \
        'latitude':mat[:,2], 'longitude':mat[:,3]}
    data['latitude'] = pd.to_numeric(data['latitude'])
    data['longitude'] = pd.to_numeric(data['longitude'])
    df_data = df(data, columns=['shop_name', 'brand', 'longitude', 'latitude'])
    return df_data
```

```python
def populate_binfo(p_d, s_i):
    bargain = []
    p_n = []
    s_n = []
    price = []
    p_c = []
    p_t = []
    end_time = []
    orig_p = []
    discount_rate = []
    brand = []
    pack_price = []
    price_100g = []
    pack_ul = []
    pack_ll = []
    for i in range(len(p_d['product_name'])):
        name = p_d['product_name'][i]
        arr = [p_d['pack_c'][i], p_d['pack_nc'][i], p_d['single_item'][i], p_d['weight'][i], s_i['shop_name']]
        p_b_lst = create_product(name, arr)
        dummy = [bargain.append(row) for row in p_b_lst]
    mat = np.array(bargain)
    data = {'product_name': mat[:,0], 'shop_name': mat[:,1], 'price': mat[:,2], \
            'end_time': mat[:,3], 'price_peritem': mat[:,4], 'pack_ll':mat[:,5], \
            'pack_ul': mat[:,6], 'pack_size': mat[:,7], 'item_size': mat[:,8], 'pack_type':mat[:,9]}
    df_data = df(data, columns=['product_name', 'shop_name', 'price', 'end_time', \
            'price_peritem', 'pack_ll', 'pack_ul', 'pack_size', 'item_size', 'pack_type'])
    return df_data
```

```python
def create_product(name, data):
    pn = 25
    d_price = [80, 100, 120, 140]
    d_ncp_price = 200
    d_pstep = 100
    d_step = 25
    exp_date = datetime.datetime(2019, 8, 18)
    product_lst = []
    p_c = data[0]
    p_nc = data[1]
    si = data[2]
    w = data[3]
    size = ['S', 'M', 'L']
    ic = range(2, 5)
    prob = [0.25, 0.5, 0.25]
    ul_p = [0.25, 0.5, 0.25]
    if(p_c):
        n = 3 * pn
        for i in range(n):
            product = [None] * 12
            product[0] = name
            product[1] = random.choice(data[4])
            product[3] = exp_date
            ll = np.random.choice(ic, p=prob)
            ul = ll + np.random.choice([0, 1, 2], p=ul_p)
            product[5] = ll
            product[6] = ul
            item_size = np.random.choice(size, p = ul_p)
            product[8] = item_size
            price_pi = random.choice(d_price) + (size.index(item_size) - 1) * d_step + random.choice(range(-20, 21, 1))
            product[4] = price_pi
            price = int(price_pi * float(ul + ll)/2)
            product[2] = price
            product_lst.append(product)
            product[9] = 'pack_countable'
    if(p_nc):
        n =  3 * pn
        for i in range(n):
            product = [None] * 12
            product[0] = name
            product[1] = random.choice(data[4])
            product[3] = exp_date
            pack_size = np.random.choice(size, p = ul_p)
            item_size = np.random.choice(size, p = ul_p)
            product[7] = pack_size
            product[8] = item_size
            price = d_ncp_price + (size.index(pack_size) - 1) * 100 + random.choice(range(-20, 21, 1))
            product[2] = price
            product_lst.append(product)
            product[9] = 'pack_notcountable'
    if(si):
        n = 3 * pn
        for i in range(n):
            product = [None] * 12
            product[0] = name
            product[1] = random.choice(data[4])
            product[3] = exp_date
            item_size = np.random.choice(size, p = ul_p)
            product[8] = item_size
            price = random.choice(d_price) + (size.index(item_size) - 1) * d_step + random.choice(range(-20, 21, 1))
            product[2] = price
            product[4] = price
            product_lst.append(product)
            product[9] = 'single_item'
    if(w):
        n = pn
        for i in range(n):
            product = [None] * 12
            product[0] = name
            product[1] = random.choice(data[4])
            product[3] = exp_date
            price = random.choice(d_price) + random.choice(range(-20, 21, 1))
            pack_size = np.random.choice(size, p = ul_p)
            product[7] = pack_size
            product[2] = price
            product[9] = 'pack_weight'
            product_lst.append(product)
    return product_lst
```
